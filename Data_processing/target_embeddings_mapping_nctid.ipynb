```python
def druglist2emb(drugids, drug2emb):
  '''
  Retrieving a single embedding from a list of drugs. 

  Input: 
    drugids (list): list with drug ids
    drug2emb (dict): dict of drug to embeddings

  Returns:
    meanembed (tensor): tensor with shape ([1024]) with embeddings of list of drugs
  '''
  embeds = []
  for drug in drugids:
    if (drug != 'none') & (drug != 'placebo') & (drug in drug2emb.keys()):
      drugemb = torch.from_numpy(drug2emb[drug])
    else: 
      drugemb = torch.zeros(1024)

    embeds.append(drugemb)

  meanembed = embeds[0]
  for i in range(1, len(embeds)): 
    meanembed = torch.add(meanembed, embeds[i])

  meanembed = meanembed / len(embeds)

  return meanembed
```

```python
df_temp = train[['nctid', 'drugbank_id']]
dict_crit = df_temp.set_index('nctid').T.to_dict('list')

nctid2target = dict()

for key, value in dict_crit.items(): 
  nctid2target[key] = druglist2emb(value[0], target_embs)
```
